# Covid Cases differentiated in Graphs of Maharashtra State

```python
#import all packages
import csv
import datetime
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
from scipy.interpolate import interp1d
import numpy as np

figure(num=None, figsize=(29, 27), dpi=200, facecolor='w', edgecolor='k')
#add csv file 
f = open('covid_19_india.csv')
csv_f = csv.reader(f)
state = "Maharashtra"
result = []
for row in csv_f:
    if state in row[3]:
        result.append(row)

date_format = '%d/%m/%y'  # The format

min_date = datetime.datetime(2020, 6, 1)
max_date = datetime.datetime(2020, 6, 30)

output = []
for row in result:
    row_date = datetime.datetime.strptime(row[1], date_format)
    if row_date >= min_date and row_date <= max_date:
        output.append(row)
dt=[]
inf=[]
conf=[]
recov=[]
for o in output:
    print("------------------------------------------------------------------")
    ans = o[0]+"    "+o[1]+"   "+o[2]+"   "+o[3] + \
        "  "+o[5]+"  "+o[6]+"  "+o[7]+"  "+o[8]
    print(ans)
    dt.append(o[1])
    inf.append(o[7])
    conf.append(o[8])
    recov.append(o[6])
```

```python
figure(num=None, figsize=(29, 27), dpi=200, facecolor='w', edgecolor='k')
#plot graph
x=dt
y=inf
x1=dt
y1=recov
x2=dt
y2=conf
plt.plot(x, y,color='red', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12) 
plt.plot(x1, y1,color='green', linestyle='dashed', linewidth = 1, 
         marker='p', markerfacecolor='blue', markersize=12) 
plt.plot(x2, y2,color='blue', linestyle='dashed', linewidth = 1, 
         marker='o', markerfacecolor='blue', markersize=12)  
plt.xlabel('Date') 
plt.ylabel('Infected & Recovered') 
plt.title('Graph on Covid-19 Data of Infected cases, Recovered Cases & Death Cases in Maharashtra')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
  
# function to show the plot 
plt.show() 
```
